```python
%load_ext autoreload
%autoreload 2
import numpy as np
import matplotlib.pyplot as plt
import h5py
import fitsio
```

```python
input_loc = '/bsuhome/hwu/scratch/uchuu/MiniUchuu/'
data = h5py.File(input_loc+f'MiniUchuu_halolist_z0p30.h5', 'r')
hid = np.array(data['id'])
pid = np.array(data['pid'])
sel1 = (pid == -1)
M200m = np.array(data['M200b'])[sel1]
sel2 = M200m > 1e11
M200m = M200m[sel2]
x_halo = np.array(data['x'])[sel1][sel2]
y_halo = np.array(data['y'])[sel1][sel2]
z_halo = np.array(data['z'])[sel1][sel2]
```

```python
nhalo = len(x_halo)
print('nhalo = ', nhalo)
rhocrit = 2.775e11 # h-unit
Om0 = 0.3089
R200m = (3 * M200m / (4 * np.pi * rhocrit * Om0 * 200.))**(1./3.) # cMpc/h (chimp)
#print(max(R200m))
```

## galaxy catalog

```python
loc = '/bsuhome/hwu/scratch/hod-selection-bias/output_mini_uchuu/model_0/'
fname = loc + 'gals.fit'
data, header = fitsio.read(fname, header=True)
x_gal = data['px']
y_gal = data['py']
z_gal = data['pz']
sel = (z_gal < 10)
plt.scatter(x_gal[sel], y_gal[sel], s=0.1, alpha=0.1)
```

```python
from scipy import spatial
rmax_tree = 4
```

```python
%%time
gal_position = np.dstack([x_gal, y_gal, z_gal])[0]
gal_tree = spatial.cKDTree(gal_position)

halo_position = np.dstack([x_halo, y_halo, z_halo])[0]
halo_tree = spatial.cKDTree(halo_position)
```

```python
%%time
indexes_tree = halo_tree.query_ball_tree(gal_tree, r=rmax_tree)
```

```python
%%time
ngal = []
for i_halo in range(nhalo):
    gal_ind = indexes_tree[i_halo]
    x_cen = x_halo[i_halo]
    y_cen = y_halo[i_halo]
    z_cen = z_halo[i_halo]
    r = (x_gal[gal_ind] - x_cen)**2 + (y_gal[gal_ind] - y_cen)**2 + (z_gal[gal_ind] - z_cen)**2 
    r = np.sqrt(r)
    ngal.append(len(r[r <= R200m[i_halo]*(1+1e-4)]))
```

```python
data = np.array([M200m, ngal]).transpose()
fname = loc + 'temp/ngal_M200m_recovered.dat'
np.savetxt(fname, data, header='M200m, ngal', fmt='%-12e %-12i')
```

```python
%load_ext autoreload
%autoreload 2
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
loc = '/bsuhome/hwu/scratch/hod-selection-bias/output_mini_uchuu/model0'
fname = loc+'temp/ngal_M200m_recovered.dat'
data = pd.read_csv(fname, delim_whitespace=True, dtype=np.float64,
                   comment='#', 
                   names=['M200m', 'ngal'])
```

```python
print('max M = %e'%(max(data['M200m'])))
```

#### focusing on the central transition region

```python
%%time
M = data['M200m']
ngal = data['ngal']
sel = (M < 1e12)
x = np.array(np.log(data['M200m']))[sel][::100]
y = np.array(data['ngal'])[sel][::100]
from kllr import kllr_model
lm = kllr_model(kernel_type='gaussian', kernel_width=0.2)
xrange, yrange_mean, intercept, slope, scatter, skew, kurt = lm.fit(x, y, bins=21)                                   

intercept_mean = np.mean(intercept, axis=0)
slope_mean = np.mean(slope, axis=0)
scatter_mean = np.mean(scatter, axis=0)
y_mean = np.mean(yrange_mean, axis=0)

plt.plot(np.exp(xrange), y_mean, label='mini uchuu recovered')

## INPUT HOD
import sys
sys.path.append('../hod')
from fid_hod import Ngal_S20_noscatt_noround
M200m_plot_list = 10**np.linspace(11, 12, 10000)
N_plot_list = []
for M200m_plot in M200m_plot_list:
    N_plot_list.append(Ngal_S20_noscatt_noround(M200m_plot))
plt.plot(M200m_plot_list, N_plot_list, c='C1', lw=2, label='input')
plt.axvline(10**11.7, c='gray', ls='--', label=r'$\rm M_{cut}$')
plt.xscale('log')
plt.legend()
plt.xlabel(r'$\rm M_{200m}$')
plt.ylabel(r'$\rm \langle N_{gal} \rangle$')
#plt.yscale('log')
plt.savefig('../../plots/uchuu/validate_gal_cat_mini_uchuu_low_M.png')
```
