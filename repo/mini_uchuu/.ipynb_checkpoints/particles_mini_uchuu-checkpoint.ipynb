```python
import h5py
input_loc = '/bsuhome/hwu/scratch/uchuu/MiniUchuu/'
data = h5py.File(input_loc+f'MiniUchuu_halolist_z0p30.h5', 'r')
hid = np.array(data['id'])
pid = np.array(data['pid'])
sel1 = (pid == -1)
M200m = np.array(data['M200b'])[sel1]
sel2 = M200m > 1e14
x_h = np.array(data['x'])[sel1][sel2]
y_h = np.array(data['y'])[sel1][sel2]
z_h = np.array(data['z'])[sel1][sel2]
mean_M = np.mean(M200m[sel2])
```

```python
data = fitsio.read(loc+'particles_043_0.1percent.fit')
x_p = data['x']
y_p = data['y']
z_p = data['z']
mpart = snap_header.mass[1]*1e10 * 1000 # 0.1% subsampling
rp_1000, Sigma_1000, DeltaSigma_1000 = measure_lensing(x_h, y_h, z_h, x_p, y_p, z_p, boxsize=400, mpart=mpart)
```

```python
data = fitsio.read(loc+'particles_043_1percent.fit')
x_p = data['x']
y_p = data['y']
z_p = data['z']
mpart = snap_header.mass[1]*1e10 * 100 # 1% subsampling
rp_100, Sigma_100, DeltaSigma_100 = measure_lensing(x_h, y_h, z_h, x_p, y_p, z_p, boxsize=400, mpart=mpart)
```

```python
plt.semilogx(rp_100, DeltaSigma_100/DeltaSigma_1000)
plt.xlabel(r'r_p')
plt.ylabel(r'$\Delta\Sigma ratio$')
```

```python
plt.loglog(rp_1000, DeltaSigma_1000, label='mini uchuu 0.1\%')
plt.loglog(rp_100, DeltaSigma_100, 'o', label='mini uchuu 1\%')

from colossus.cosmology import cosmology
from colossus.halo import profile_nfw

#http://skiesanduniverses.org/Simulations/Uchuu/
params = {'flat': True, 'H0': 67.74, 'Om0': 0.3089, 'Ob0': 0.0486, 'sigma8': 0.8159, 'ns': 0.9667}
cosmology.addCosmology('Uchuu', params)
cosmo = cosmology.setCosmology('Uchuu')

redshift = 0.3
profile = profile_nfw.NFWProfile(M=mean_M, mdef='200m', z=redshift, c=5)
r = 10**np.linspace(1.5, 4.2, 100) 
Sigma = profile.surfaceDensity(r)
DeltaSigma = profile.deltaSigma(r)

a = 1/(1+redshift)
plt.loglog(r * 1e-3 / a, DeltaSigma * 1e6 * a**2, label='colossus NFW')
plt.xlabel(r'$\rm r_p [cMpc/h]$')
plt.ylabel(r'$\rm \Delta\Sigma\ [h M_\odot / cMpc^2]$')

plt.legend()
```
